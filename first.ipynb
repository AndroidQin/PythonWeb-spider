### 伪装浏览器

```python
from urllib import request
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36"}#请求头信息
req = request.Request("https://edu.csdn.net/course/detail/6447",headers=headers)#request类，更加高级的网址访问
resp = request.urlopen(req)#接收request类中的内容
print(resp.read().decode())#读取+解码
```

## cookiejar模块

```python
from urllib import request
from urllib import parse
from http.cookiejar import CookieJar


headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.119 Safari/537.36"
}
#d登陆
#1.1创建Cookiejar对象
cookiejar = CookieJar()
#1.2使用Cookiejar创建一个HTTPCookieProcessor对象
handler = request.HTTPCookieProcessor(cookiejar)
#1.3使用上一步的创建的handler创建一个opener
opener = request.build_opener(handler)
#1.4使用opener发送登陆请求
post_url = 'https://i.meishi.cc/login.php?redirect=https%3A%2F%2Fwww.meishij.net%2F'
post_data = parse.urlencode({#转码
    'username':'1097566154@qq.com',
    'password':'wq15290884759.'
})
req = request.Request(post_url,data=post_data.encode("utf-8"))
opener.open(req)

#2.访问个人网页
url = 'https://i.meishi.cc/cook.php?id=13686422'
rq = request.Request(url,headers=headers)
resp = opener.open(rq)
print(resp.read().decode())
```
